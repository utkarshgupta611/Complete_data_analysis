let's check the skewness of alchohal in wine

```python
from scipy.stats import skew
print('skew of white wine ', skew(df_white['alcohol']))
print('skew of red wine ', skew(df_red['alcohol']))
```

```python
def draw_hist(temp_df, bin_size=15):
    ax =sns.distplot(temp_df)
    plt.tight_layout()
    plt.locator_params(axis='y', nbin = 6)
    plt.show()
    print('skewness is {}'.format(skew(temp_df)))
    print('mean is {}'.format(np.mean(temp_df)))
    print('median is {}'.format(np.median(temp_df)))
```

```python
draw_hist(df_white['alcohol'])
```

```python
draw_hist(df_red['alcohol'])
```

```python
from scipy.stats import pearsonr
def get_corr(col1, col2, temp_df):
    pearson_corr,p_value = pearsonr(temp_df[col1],temp_df[col2])
    print("Correlation between {} and {} is {}".format(col1, col2, pearson_corr))
    print("P-value of this correlation is {}".format(p_value))
```

```python
get_corr('alcohol', 'pH', df_white)
```

```python
get_corr('alcohol', 'pH', df_red)
```
